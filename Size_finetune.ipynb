```python
import os
import gc
import torch
import pandas as pd
from ultralytics import YOLO

# 1. 기본 경로 및 평가할 해상도 리스트 정의
BASE_DIR = "/Users/gyuminkang/Desktop/YOLO/runs/detect/runs/tune_cells"
RESOLUTIONS = [320, 480, 640, 768, 896, 1024]
DATA_YAML = "data.yaml"  # 데이터셋 설정 파일 이름

results_list = []

print("🏁 [종합 평가] 각 해상도별 'best.pt' 탐색 및 Validation(검증셋) 채점 시작...\n")

for size in RESOLUTIONS:
    # 경로 구조 반영
    model_path = os.path.join(BASE_DIR, f"imgsz_{size}", "weights", "best.pt")
    
    # 해당 해상도의 가중치 파일이 존재하는지 확인
    if not os.path.exists(model_path):
        print(f"⚠️ [imgsz_{size}] 가중치 파일이 없어 건너뜁니다. (패스: {model_path})")
        continue
        
    print(f"🔍 [imgsz_{size}] 가중치 로드 중... 검증(Validation) 평가를 시작합니다.")
    
    try:
        # 모델 객체 생성
        model = YOLO(model_path)
        
        # 🌟 핵심 수정: Test가 아닌 Validation 데이터셋으로 평가 실행
        metrics = model.val(
            data=DATA_YAML,
            split='val',    # 테스트 셋이 아닌 검증(Validation) 데이터 사용
            imgsz=size,     # 해당 가중치가 학습했던 해상도 그대로 평가
            device='mps',   # Mac GPU 가속 적용
            batch=8,
            plots=False     # 평가 속도를 높이기 위해 불필요한 이미지 저장은 비활성화
        )
        
        # 핵심 지표 수집 및 반올림 처리
        results_list.append({
            "해상도 (imgsz)": f"imgsz_{size}",
            "mAP50": round(metrics.box.map50, 4),
            "mAP50-95": round(metrics.box.map, 4),
            "정밀도 (Precision)": round(metrics.box.mp, 4),
            "재현율 (Recall)": round(metrics.box.mr, 4),
            "추론속도 (ms)": round(metrics.speed['inference'], 2)
        })
        
        # 메모리 누수 방지를 위한 즉시 해제
        del model
        gc.collect()
        if torch.backends.mps.is_available():
            torch.mps.empty_cache()
            
    except Exception as e:
        print(f"❌ [imgsz_{size}] 평가 중 오류 발생: {e}")
        print("-" * 50)

# 2. 모든 평가 완료 후 통합 성적표 작성 및 순위 정렬
print("\n" + "="*70)
print("🏆 [튜닝 결과] Validation(검증셋) 기준 해상도별 성능 순위표")
print("="*70)

if results_list:
    df = pd.DataFrame(results_list)
    
    # 핵심 지표인 mAP50 기준으로 내림차순 정렬 (1등이 가장 위로)
    df = df.sort_values(by="mAP50", ascending=False).reset_index(drop=True)
    
    # 가독성을 위해 인덱스를 1등부터 순위로 변경
    df.index = df.index + 1
    df.index.name = "순위"
    
    # 주피터 노트북에 예쁜 마크다운 테이블 형태로 출력
    print(df.to_markdown())
    print("="*70)
    print(f"🎯 최종 결론: Validation 셋에서 가장 성능이 좋은 모델은 [{df.iloc[0]['해상도 (imgsz)']}] 이며, 이를 최종 본 학습에 채택합니다!")
else:
    print("❌ 평가를 완료한 모델이 없습니다. 가중치 파일 경로를 다시 확인해 주세요.")
```

```python
import os
import gc
import torch
import pandas as pd
from ultralytics import YOLO

# 1. 기본 경로 및 평가할 해상도 리스트 정의
BASE_DIR = "/Users/gyuminkang/Desktop/YOLO/runs/detect/runs/tune_cells"
RESOLUTIONS = [320, 480, 640, 768, 896, 1024]
DATA_YAML = "data.yaml"  # 데이터셋 설정 파일 이름

results_list = []

print("🏁 [종합 평가] 각 해상도별 'best.pt' 탐색 및 Test 데이터셋 채점 시작...\n")

for size in RESOLUTIONS:
    # 사용자님이 제공하신 경로 구조 반영
    model_path = os.path.join(BASE_DIR, f"imgsz_{size}", "weights", "best.pt")
    
    # 해당 해상도의 가중치 파일이 존재하는지 확인
    if not os.path.exists(model_path):
        print(f"⚠️ [imgsz_{size}] 가중치 파일이 없어 건너뜁니다. (패스: {model_path})")
        continue
        
    print(f"🔍 [imgsz_{size}] 가중치 로드 중... 최종 평가를 시작합니다.")
    
    try:
        # 모델 객체 생성
        model = YOLO(model_path)
        
        # 순수한 Test 데이터셋으로 평가 실행
        metrics = model.val(
            data=DATA_YAML,
            split='test',   # 벨리데이션이 아닌 완전히 분리된 테스트 데이터 사용
            imgsz=size,     # 해당 가중치가 학습했던 해상도 그대로 평가
            device='mps',   # Mac GPU 가속 적용
            batch=8,
            plots=False     # 평가 속도를 높이기 위해 불필요한 이미지 저장은 비활성화
        )
        
        # 핵심 지표 수집 및 반올림 처리
        results_list.append({
            "해상도 (imgsz)": f"imgsz_{size}",
            "mAP50": round(metrics.box.map50, 4),
            "mAP50-95": round(metrics.box.map, 4),
            "정밀도 (Precision)": round(metrics.box.mp, 4),
            "재현율 (Recall)": round(metrics.box.mr, 4),
            "추론속도 (ms)": round(metrics.speed['inference'], 2)
        })
        
        # 메모리 누수 방지를 위한 즉시 해제
        del model
        gc.collect()
        if torch.backends.mps.is_available():
            torch.mps.empty_cache()
            
    except Exception as e:
        print(f"❌ [imgsz_{size}] 평가 중 오류 발생: {e}")
        print("-" * 50)

# 2. 모든 평가 완료 후 통합 성적표 작성 및 순위 정렬
print("\n" + "="*70)
print("🏆 [최종 결과] 테스트셋 기준 해상도별 성능 순위표")
print("="*70)

if results_list:
    df = pd.DataFrame(results_list)
    
    # 핵심 지표인 mAP50 기준으로 내림차순 정렬 (1등이 가장 위로)
    df = df.sort_values(by="mAP50", ascending=False).reset_index(drop=True)
    
    # 가독성을 위해 인덱스를 1등부터 순위로 변경
    df.index = df.index + 1
    df.index.name = "순위"
    
    # 주피터 노트북에 예쁜 마크다운 테이블 형태로 출력
    print(df.to_markdown())
    print("="*70)
    print(f"🎯 최종 결론: 테스트셋에서 가장 성능이 좋은 모델은 [{df.iloc[0]['해상도 (imgsz)']}] 입니다!")
else:
    print("❌ 평가를 완료한 모델이 없습니다. 가중치 파일 경로를 다시 확인해 주세요.")
```
